```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
#Relação entre a variável de entrada (area_m2) e a variavel alvo (valor_alguel)
sns.scatterplot(data = df, x = 'area_m2', y = 'valor_aluguel')
```

```python
#Construão do Modelo OLS

#Definrir a variavel dependente
y = df['valor_aluguel']

#Definir  a variavel independente
x = df['area_m2']
```

```python
#Adição de uma constante à variável independente
x = sm.add_constant(x)
```

```python
#Criação do modelo
modelo = sm.OLS(y, x)

#Treunamento do Modelo
resultado = modelo.fit()
print(resultado.summary())
```

```python
#Plot
plt.figure(figsize = (12, 8))
plt.xlabel('area_m2', size = 16)
plt.ylabel('valor_aluguel', size = 16)
plt.plot(x['area_m2'], y, 'o', label = 'Dados Reais')
plt.plot(x['area_m2'], resultado.fittedvalues, 'r-', label = 'Linha de Regressão (Previsões do Modelo)')
plt.legend(loc = 'best')
plt.show()
```
